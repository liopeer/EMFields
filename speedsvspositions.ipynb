```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from matplotlib import cm
```

```python
data = np.genfromtxt("speedsvspositions.csv", delimiter=',')
speeds = np.array([30,50,80])
proximity = data[:,-1]
```

```python
data[0,0]=90.3
```

```python
speed, prox = np.meshgrid(speeds, proximity)
exp = np.transpose(np.asarray([data[:,1], data[:,3], data[:,5]]))
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')
ax.scatter(speed, prox, exp)
```

```python
fig, ax = plt.subplots(subplot_kw={"projection":"3d"}, figsize=(10,10))
ax.plot_surface(speed, prox, exp, cmap=cm.coolwarm)
ax.set_xlabel("speeds")
ax.set_ylabel("proximity")
ax.set_zlabel("exposure")
```
