## Plotting the observed frequencies vs speeds

```python
speeds = np.array([30, 40, 50, 80])
freqs = np.array([107, 143, 178, 286])

lin_model = np.polyfit(speeds, freqs, 1)
slope = lin_model[0]
intercept = lin_model[1]

x = np.arange(100)
y = slope*x + intercept

plt.figure()
plt.axis([0,100,0,300])
plt.scatter(speeds, freqs, color='red')
plt.title("Speeds vs. Frequencies from rear motor")
plt.xlabel("Speed [km/h]")
plt.ylabel("Dominant frequency [Hz]")
plt.annotate("intercept at " + str(round(intercept, 2)), (0 + 13, 0 + 20))
plt.plot(x,y)
#plt.savefig("freqvsspeed.png")
plt.show()
```

## Setting Motor revolution (rpm) in relation to frequency

```python
wheel_diameter = 0.67
dist_per_r = np.pi * wheel_diameter
speeds = np.array([30,40,50,80])

rpm = speeds/3.6 / dist_per_r
```

```python
speeds = np.array([30, 40, 50, 80])
freqs = np.array([107, 143, 178, 286])

lin_model = np.polyfit(speeds, freqs, 1)
slope = lin_model[0]
intercept = lin_model[1]

x = np.arange(100)
y = slope*x + intercept

plt.figure()
plt.axis([0,100,0,300])
plt.scatter(speeds, freqs, color='red')
plt.title("Speeds vs. Frequencies from rear motor")
plt.xlabel("Speed [km/h]")
plt.ylabel("Dominant frequency [Hz]")
plt.annotate("intercept at " + str(round(intercept, 2)), (0 + 13, 0 + 20))
plt.plot(x,y)
plt.plot(speeds, rpm*9*3)
#plt.savefig("freqvsspeedvsrpm.png")
plt.show()
```
