```python
import numpy as np
import matplotlib.pyplot as plt
```

# Data Import
## Import time domain CSV file

```python
def import_time(file):
    """
    INPUT: path to TD type file (str)
    
    OUTPUT: time domain np array
    # 0: time
    # 1: x-axis [muT]
    # 2: y-axis
    # 3: z-axis
    # 4: magn
    """
    time_domain = np.genfromtxt(file, delimiter='\t', skip_header=True)
    time_domain = time_domain[:,:-4]
    
    return time_domain
    
file = "/Users/lionelpeer/polybox/ETH/HS21/EMFelderimAlltag/Messungen/Messungen1/50kmh/rueckbank_mitte_TD.csv"

time_d = import_time(file)
```

## Import frequency domain CSV file

```python
def import_spec(file):
    """
    INPUT: path to SPEC type file (str)
    
    OUTPUT: spec domain np array
    # 0: frequency [Hz]
    # 1: x-axis [muT]
    # 2: y-axis
    # 3: z-axis
    # 4: magn
    """
    spec_domain = np.genfromtxt(file, delimiter='\t', skip_header=True)
    return spec_domain


path = "/Users/lionelpeer/polybox/ETH/HS21/EMFelderimAlltag/Messungen/Messungen1/50kmh/rueckbank_mitte_SPEC.csv"

spec_d = import_spec(path)
```

# Messungen 1
## 40 km/h

```python
path = "/Users/lionelpeer/polybox/ETH/HS21/EMFelderimAlltag/Messungen/Messungen1/40kmh/"
t_1 = import_time(path + "rueckbank_mitte_TD.csv")
t_2 = import_time(path + "rueckbank_mitte2_TD.csv")

s_1 = import_spec(path + "rueckbank_mitte_SPEC.csv")
s_2 = import_spec(path + "rueckbank_mitte2_SPEC.csv")
```

## Plot frequency domain

```python
WOI = np.s_[10:] # frequency window of interest: DC fields carry too much signal
max_index = np.argmax(s_1[WOI, 4])
maximum = np.max(s_1[WOI, 4])

plt.figure()
plt.title("40 km/h spectrum")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Field strength [μT]")
marginx = 40
plt.annotate("max of " + str(round(maximum, 2)) + "μT at " + str(max_index) + "Hz", (max_index + marginx, maximum))
plt.plot(s_1[WOI,0], s_1[WOI,4], color='black', marker='.', markerfacecolor='red', markersize=15)
```

## Plot time domain

```python
TOI = np.s_[:]

plt.figure(figsize=(30,5))
plt.title("40 km/h time domain")
plt.xlabel("Time [ms]")
plt.ylabel("Field strength [μT]")
plt.plot(t_2[TOI,0]*1e3, t_2[TOI,4], color="black")
```

comparing mean value of time domain with the DC part from the spectrum; should be roughly equal

```python
mean = np.mean(t_2[:,4])
dc_part = s_2[0,4]

print("mean of time domain:", mean, "\ndc part of from frequency domain", dc_part)
```

which indeed they are

let's plot de-meaned/without DC

```python
TOI = np.s_[:]

plt.figure(figsize=(30,5))
plt.title("40 km/h time domain")
plt.xlabel("Time [ms]")
plt.ylabel("Field strength [μT]")
plt.plot(t_2[TOI,0]*1e3, t_2[TOI,4] - dc_part, color="black")
```

let's try to just mark the DC part

```python
TOI = np.s_[:]

plt.figure(figsize=(30,5))
plt.title("40 km/h time domain")
plt.xlabel("Time [ms]")
plt.ylabel("Field strength [μT]")
plt.plot(t_2[TOI,0]*1e3, t_2[TOI,4], color="black")
plt.plot(t_2[TOI,0]*1e3, np.asarray([dc_part for i in np.arange(t_2[TOI,4].shape[0])]), color="blue")
```
